## La Liga
Founded in 1929, La Liga, or Spanish Primera Division is the top division of Spanish football. It contains 20 teams. The season runs from August to May, and teams play each other both home and away to fulfil a total of 38 games. La Liga has promotion and relegation linked to the Spanish Segunda Division, the second tier. Three teams are relegation from La Liga, and three teams are promoted from the Championship, each season. Sixty-two different clubs have competed since its inception, with Real Madrid winning the most titles. The competition's all-time leading goalscorer is Lionel Messi.

#### Total Goals Scored Per Team from 2012 - 2022
Barcelona have the highest goals followed by Real Madrid and Atletico Madrid. Cordoba have the lowest goals scored. This is to be expected since they have only featured in La Liga just once (2014/2015) since 1971/1972

```python
Goals = LaLiga.groupby(['Team'])[["GF"]].sum()

plt.figure(figsize=(20,20))
plt.barh(Goals.index, Goals['GF'], color = "green", label=LaLiga["GA"], height=0.5)
plt.xlabel("Team", fontsize = 18)
plt.ylabel("Number of Goals", fontsize = 18)
plt.title("Number of Goals Scored Per Team 2012 - 2022", fontsize = 20)
plt.yticks(fontsize=16)
plt.show()                               
```

#### Total Scores Every Year Since 2012

The highest number of goals scored was reached in 2016. After 2016, there has been a decrease in the total number of goals scored collectively by all the teams. 2022 has the lowest total score so far and this is mainly because the league is still ongoing. 

```python
Yearly_Goals= LaLiga.groupby('Season', as_index=False)[['GF']].sum()

plt.figure(figsize=(5,10))
plt.bar(Yearly_Goals['Season'], Yearly_Goals['GF'], width=0.7, color = ['r','g','b', 'y','m','c','k'], hatch = ("*"))
plt.xlabel("Year", fontsize = 18)
plt.ylabel("Number of Goals", fontsize = 18)
plt.title("Yearly Goals 2012 - 2022", fontsize = 20)
plt.xticks()
```

#### Home and Away Goals
According to the saying, "No place like home", so as expected, all the teams scored more goals in their home matches compared to away matches.

```python
Home_Away = LaLiga.groupby(['Team', 'Venue'], as_index=False)[['GF']].sum()
sns.set(style='white')
sns.set_style('ticks')
plt.figure(figsize=(20,5))
sns.barplot(x='Team', y='GF', hue='Venue', data=Home_Away,
            palette=['purple', 'brown'])
plt.title('Volume of Goal Home and Away 2012-2022', fontsize=20)
plt.xlabel('Team', fontsize=18)
plt.ylabel('Number of Goals', fontsize=18)
plt.xticks(rotation=80, fontsize=15)
plt.show()
```

#### Average Total Shots By Teams Per Season

```python
Shots = LaLiga.groupby('Season', as_index=False)[['Sh']].mean()
Shots['Season']= Shots[Shots['Season'] >= 2015]
Shots['Season']= Shots[Shots['Season'] < 2022]

plt.figure(figsize=(10,5))
sns.set(style='white')
sns.despine()
sns.set_style('ticks')

plt.scatter(data=Shots, x= 'Season', y= 'Sh')
plt.plot(Shots['Season'], Shots['Sh'])
plt.title('Average Total Shots By Teams Per Season', fontsize=16)
plt.xlabel('Season')
plt.ylabel('Average Shots')
plt.xticks(rotation=45)

plt.show()
```

#### Highest Number of Goals Recorded in a Single Match by Each Team
Real Madrid has the highest number of goals recorded in a single match (Highest Goal =10)
All the teams at some point between 2012 - 2022 scored more than 1 goal in a single match

```python
Highest_Goals_Per_Match = LaLiga.groupby("Team", as_index=False)[['GF']].max()

plt.figure(figsize=(20,10));
plt.bar(Highest_Goals_Per_Match['Team'], Highest_Goals_Per_Match['GF'], width=0.7, color = ['r','g','b', 'y','m','c','k']);
plt.xlabel("Team", fontsize = 18);
plt.ylabel("Goals Scored", fontsize = 18);
plt.title("Highest Numbers of Goals Scored in a Match", fontsize = 20);
plt.xticks(rotation=80, fontsize = 15);
```
